## Exercise 1: Visualizing Data Over Time

1. Extract the data and country names from the CSV file as we did in class. Call the country names `all_countries`.

```julia
using CSV
using DataFrames
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
download(url, "covid_data.csv")
data = CSV.read("covid_data.csv")
rename!(data, 1 => "province", 2 => "country") # ! is convention: function *modifies* its argument in place
countries_col = collect(data[:, 2])
all_countries = unique(countries_col);
```

2. Make a `Vector` called `countries` with a subset of countries that we wish to plot, say China, Japan, South Korea, US, United Kingdom, France, Germany. Be careful to check how they are written in the data set.

```julia
countries = vec(["China" "Japan" "Korea, South" "US" "Spain" "France" "Germany"]);
```

3. Define a variable `num_days` by extracting the number of days of data from the dataframe.

```julia
num_days = ncol(data) - 4
```

4. We need to accumulate the data for those places that are split up into territories. Make a zero vector of the correct length for each country, e.g. using the function `zeros`. Loop through all the countries and add the corresponding data to that country’s data. You may use a dictionary (`Dict`), or a matrix, or a `Vector` containing `Vectors`, or a new `DataFrame` to store the data.

```julia
country_time_series = Dict{String,Array{Real,1}}()
for country in countries
    num_territories = count(data.country .== country)
    first_index = findfirst(data.country .== country)
    country_slice = data[first_index:first_index+num_territories-1, 5:end]
    infection_time_series = vec([sum(country_slice[:,i]) for i in 1:size(country_slice,2)])
    push!(country_time_series, country => infection_time_series)
end
```

5. Plot all countries’ data on the same graph by using a `for` loop. To do so, first make an empty plot: <br> `p = plot()`<br> Then run the loop, adding in the data using `plot!`<br> Finally, display the plot by evaluating the plot object `p`. (Just type its name and evaluate.)

```julia
using Plots
using Dates
df = DateFormat("m/d/y")
dates = [Date(string(i),df) for i in names(data)[5:end]]
dates.+= Dates.Year(2000)
p = plot(leg=:topleft, xticks = dates[1:5:end], xrotation = 45)
for i in 1:length(countries)
    plot!(map(Date,dates),country_time_series[countries[i]], label = countries[i])
end
xlabel!("Date")
ylabel!("Number of Infections")
p
```

6. Now use a log scale on the $y$ axis. In order to do so you will need to convert the vectors to contain `Float64` and replace any 0 values by NaN (“not a number”) so that `Plots.jl` ignores those values. <br> Is there exponential growth?

```julia
df = DateFormat("m/d/y")
replace_zero(vec) = map(x -> iszero(x) ? NaN : x , vec)
dates = [Date(string(i),df) for i in names(data)[5:end]]
dates.+= Dates.Year(2000)
p = plot(leg=:bottomright, xticks = dates[1:5:end], xrotation = 45,yscale=:log10)
for i in 1:length(countries)
    plot!(map(Date,dates),replace_zero(country_time_series[countries[i]]), label = countries[i])
end
xlabel!("Date")
ylabel!("Number of Infections")
p
```

7. Turn this into an interactive visualization by adding a slider corresponding
to the current day, varying between 1 and the total number of days for
which you have data. You should draw only the data up to that particular
day. As you move the slider the plot should update. Fix the horizontal
axis using the keyword argument `xlim=(0, num_days)` inside the first `plot`
command.

```julia
using Interact
@manipulate for T in 1:num_days
    p = plot(leg=:bottomright, xrotation = 45,yscale=:log10,xticks = dates[1:5:end],)
    for i in 1:length(countries)
        plot!(map(Date,dates[1:T]), replace_zero(country_time_series[countries[i]][1:T]),  label = countries[i], m=:dot)
    end
    xlabel!("Date")
    ylabel!("Number of Infections")
    ylims!(1,600000)
    cur_xlims=(dates[1], dates[end])    # with x_data = dates[1:day_index]
    xlims!(map(Dates.value, cur_xlims))
    p
end
```

## Exercise 2: Visualizing changes

1. Make a data set `total_cases_to_date`, representing for each day the total number of confirmed cases during the whole epidemic up until that point.

```julia
total_cases_to_date = data;
```

2. Make a set of data `new_cases` which is the total number of cases only during the past 7 days. You may use the sum function.

```julia
new_cases = copy(hcat(total_cases_to_date[:,1:4],total_cases_to_date[:,12:end]))
check = copy(new_cases[:,32])
for i in 5:size(new_cases)[2] #5 because thats the first date col
    current_day_index = i
    cum_table_index = i+7
    day_change = total_cases_to_date[:,cum_table_index] .- total_cases_to_date[:,cum_table_index-7]
    new_cases[:,i] = day_change
end
```

3. Make the visualization using a slider representing days as in the previous exercise, plotting the total number of cases on the 𝑥 axis and new cases on the 𝑦 axis.You need to take care that each vector being plotted has the same length.

```julia
#First and foremost, let us redo the date ranges 
dates = [Date(string(i),df) for i in names(new_cases)[5:end]]
dates.+= Dates.Year(2000)

#then, lets work the total confirmed cases -- 12 is the magic number in this section, cuz its a week removed
total_confirmed_cases = Dict{String,Array{Real,1}}()
for country in countries
    num_territories = count(data.country .== country)
    first_index = findfirst(data.country .== country)
    country_slice = data[first_index:first_index+num_territories-1, 12:end]
    infection_time_series = vec([sum(country_slice[:,i]) for i in 1:size(country_slice,2)])
    push!(total_confirmed_cases, country => infection_time_series)
end
countries_new = copy(countries)
push!(countries_new, "All") #we also want one for everybody
all_time_series = vec([sum(data[:,i]) for i in 12:size(data,2)])
push!(total_confirmed_cases, "All" => all_time_series)

#now, same thing for cases in the last week
cases_last_week = Dict{String,Array{Real,1}}()
for country in countries
    num_territories = count(new_cases.country .== country)
    first_index = findfirst(new_cases.country .== country)
    country_slice = new_cases[first_index:first_index+num_territories-1, 5:end]
    infection_time_series = vec([sum(country_slice[:,i]) for i in 1:size(country_slice,2)])
    push!(cases_last_week, country => infection_time_series)
end
all_time_series = vec([sum(new_cases[:,i]) for i in 5:size(new_cases,2)])
push!(cases_last_week, "All" => all_time_series)

#finally, we can plot!!
using Interact
@manipulate for T in 1:length(dates)
    p = plot(leg=:topleft, xrotation = 45,yscale=:log10, xscale=:log10)
    for i in 1:length(countries_new)
        plot!(replace_zero(total_confirmed_cases[countries_new[i]][1:T]), replace_zero(cases_last_week[countries_new[i]][1:T]),  label = countries_new[i], m=:dot, markersize=:3)
    end
    xlabel!("Total confirmed cases")
    ylabel!("Cases in the Last Week")
    xlims!(5,8000000)
    ylims!(1,2000000)
    p
end
```
