# Simple Model of Infection

Each individual infects *c* new people each day

$I_{n+1} = \lambda I_n$, where $\lambda := 1 + c$ -- **growth rate**

```julia
I₀ = 1; #start with 1 infected person
```

```julia
c = 0.01; #each individual infects 0.01 new people each day (ON AVG)
```

```julia
λ = 1 + c;
```

`;` suppresses output

## Automate the process of increasing numbers

```julia
T = 1000 #num days
I = zeros(T);
```

```julia
I[1] = I₀
for x in 1:T-1
    I[x+1] = λ*I[x]
end
```

```julia
using Plots
```

```julia
plot(I, m=:o, label="I[n]", legend=:topleft)
```

## Function

```julia
function run_infection(T=20)
    I = zeros(T)
    I[1] = I₀
    for x in 1:T-1
        I[x+1] = λ*I[x]
    end
    return I
end
```

```julia
I_result = run_infection(1000);
```

```julia
plot(I_result, m=:o, label="I[n]", legend=:topleft)
```

```julia
using Interact
```

```julia
@manipulate for T in 1:300
    I_result = run_infection(T)
    plot(I_result, m=:o, label="I[n]", legend=:topleft)
    xlims!(0,T)
    ylims!(0,50)
end
```

$I_{n+1} = I_n + cI_n$

Each individual will be in contact with a fraction $\alpha$ of the population. At each contact there will be a propability $p$ that you infet each person.

So $c = p \alpha$

New model:

Can only infect uninfected people!

$I_{n+1} = I_n + [(p \alpha (N - I_n)]*I_n$


$I_{n+1} = I_n + [\beta(I_n,S_n)]*I_n $
